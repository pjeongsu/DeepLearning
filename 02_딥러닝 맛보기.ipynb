```python
# 딥러닝을 구동하는데 필요한 라이브러리

from keras.models import Sequential
from keras.layers import Dense

import pandas as pd
import tensorflow as tf
```

```python
# 수술 환자 데이터를 가져온다
# 1이 생존, 0이 사망

df1 = pd.read_csv('../data/ThoraricSurgery.csv', header=None)
df1
```

```python
# 입력과 결과로 나눈다

x = df1.drop(17, axis=1)
y = df1[17]

display(x)
display(y)
```

```python
# 딥러닝 구조를 결정한다(모델을 설정하고 실행하는 부분이다)
# 각각의 레이어들이 모델이라고 생각하면 됨
# 어떤 학습 모델을 사용할 건지 만든 거! 아직 학습을 한 것은 아님! => 딥러닝 객체를 만들고 정의

model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) # 입력을 위한 레이어 - input_dim=17:입력 데이터의 컬럼 수 
model.add(Dense(1, activation='sigmoid')) # 출력 레이어
```

```python
# 선형모델은 mean_squared_error을 사용하면 됨
```

```python
# 데이터의 랜덤성을 고정한다 - 똑같은 형태의 데이터를 가지고 학습과 평가를 할 수 있기 때문에 중요!
tf.random.set_seed(1)

# 딥러닝을 실행한다
# loss='mean_squared_error' 유사도 계산 함수를 통해서 예측값과 진짜 결과를 비교해서 얼마나 일치하는지 
# => 그 평가방법(이진 분류) : accuracy 
# => 평가 결과가 안 좋으면 수정하는 작업을 optimizer='adam' 함수 사용 - 방정식의 상수값을 보정

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

# 학습
# epochs=30 : 총 30번을 수행하겠다
# 손실 - 0.01354만큼 잘못되었다, 정확도
# 손실은 낮을수록, 정확도는 높을수록 좋다

model.fit(x, y, epochs=40, batch_size=10)
```

```python
# 예측(예측 결과가 1일 확률)
# predict_classes 는 더이상 지원 안함

pred1 = model.predict(x)
pred1
```

```python
# predict_classes 는 더이상 지원 안함
# 대신 이렇게 적어주기

result1 = (pred1 > 0.5).astype(int)
result1
```

```python
# 정확도 비교

from sklearn.metrics import accuracy_score

score1 = accuracy_score(y, result1)
score1
```
